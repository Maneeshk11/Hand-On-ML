To find the direction of increase or decrease, we first calculate the gradient. The gradient of the function is:
Î”T = [ $-200x/(x^2 + y^2 + 1)^2$, $-200y/(x^2 + y^2 + 1)^2$ ]

```python
def calculateGradient(x, y):
    return (-200*x / (x**2 + y**2 + 1)**2, -200*y / (x**2 + y**2 + 1)**2)
```

```python
greatest_increase = calculateGradient(3, 2)
greatest_increase
```

The greatest increase is in directio nof (-3.06, -2.04)

```python
def calculateMagnitude(x, y):
    return math.sqrt(x**2 + y**2)
```

```python
greatest_increase_mag = calculateMagnitude(greatest_increase[0], greatest_increase[1])
greatest_increase_mag
```

The greatest increase magnitude is calculated to be 3.67

### (6)

-> Gradient of Dot Product : $(A^T B + B^T A)x$

    we get this after considering dot product of Ax and Bx as $(Ax)^T(Bx)$ and then differentiating it w.r.t "x".

-> Hessian: $A^T B + B^T A$
    
    we get this after double differentiating $(Ax)^T(Bx)$ w.r.t. "x".


To calculate for Matrices A, B and a variable x:

```python
def gradient_dot_product(A, B, x):
    return (A.T @ B + B.T @ A) @ x

def hessian_dot_product(A, B):
    return A.T @ B + B.T @ A
```

```python
x = pd.read_csv("X.csv")
```

```python
y = pd.read_csv("y.csv")
```

```python
X = x.to_numpy()
y = y.to_numpy()
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
X_b = np.c_[np.ones((len(X), 1)), X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

Implementing the mini-batch gradient descent on the above linear regression

```python
results = []
```

#### Batch size: 1

```python
theta_path_mgd = []

m = len(X_b)

n_iterations = 50
minibatch_size = 1

np.random.seed(42)
theta = np.random.randn(2,1) 

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = 2/minibatch_size * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        theta_path_mgd.append(theta)
theta
```

```python
theta_path_mgd = np.array(theta_path_mgd)
```

```python
plt.figure(figsize=(7,4))
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.scatter(theta_path_mgd[:, 0], theta_path_mgd[:, 1])
plt.legend(loc="upper left", fontsize=12)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
plt.show()
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta)
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```
