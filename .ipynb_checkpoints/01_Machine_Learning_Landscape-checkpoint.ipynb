## Part 1 : Linear regression with a linear 

```python
dataset = pd.read_csv("lin_df.csv")
```

```python
dataset.info()
```

#### Using the result we got from the written assignment below: 

```python
X = np.array(dataset.X)
Y = np.array(dataset.Y)
```

```python
x_mean = np.mean(X)
y_mean = np.mean(Y)
x_square_mean = np.mean(X*X)
xy_mean = np.mean(X*Y)
```

```python
theta0 = (x_mean * xy_mean - x_square_mean * y_mean)/(x_mean**2 - x_square_mean)
theta1 = (x_mean * y_mean - xy_mean)/(x_mean**2 - x_square_mean)
```

```python
print("theta0 = ",theta0)
```

```python
print("theta1 = ", theta1)
```

```python
plt.scatter(dataset.X, dataset.Y)
line_x = np.linspace(0, 16)
line_y = theta0 + theta1*line_x
plt.plot(line_x, line_y, color="orange", label="generated regression line")
plt.legend()
plt.show()
```

## Assignment - 1 : Part 4

```python
xvals = np.linspace(0, 15, 100)
xtraining = nonlin_data.X.values
ytraining = nonlin_data.Y.values
k = 4
yresult = []
```

#### For K = 4

```python
for x in xvals:
    distances = np.abs(xtraining - x)
    closest_indices = np.argsort(distances)[:k]
    closest_elements = ytraining[closest_indices]
    res = np.mean(closest_elements)
    yresult.append(res)
```

```python
plt.plot(xvals, yresult, color="red")
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(nonlin_data.X, nonlin_data.Y, color="gold")
plt.show()
```

#### For K = 8

```python
yresult = []
k = 8
for x in xvals:
    distances = np.abs(xtraining - x)
    closest_indexes = np.argsort(distances)[:k]
    closest_elements = ytraining[closest_indexes]
    res = np.mean(closest_elements)
    yresult.append(res)
```

```python
plt.plot(xvals, yresult, color="red")
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(nonlin_data.X, nonlin_data.Y, color="gold")
plt.show()
```

#### For K = 16

```python
yresult = []
k = 16
for x in xvals:
    distances = np.abs(xtraining - x)
    closest_indexes = np.argsort(distances)[:k]
    closest_elements = ytraining[closest_indexes]
    res = np.mean(closest_elements)
    yresult.append(res)
```

```python
plt.plot(xvals, yresult, color="red")
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(nonlin_data.X, nonlin_data.Y, color="gold")
plt.show()
```

#### Function got smoother with increase in the value of K.


## Assignment - 1 : Part 5

```python
lindata = dataset

X = np.array(lindata.X)
Y = np.array(lindata.Y)

x_mean = np.mean(X)
y_mean = np.mean(Y)
x_sq_mean = np.mean(X*X)
xy_mean = np.mean(X*Y)
y_sq_mean = np.mean(Y*Y)
```

#### Coefficients of the Objective function can be calculated using the $\bar x$ , $\bar y$ , $\bar {x^{2}}$, $\bar {y^{2}}$ and $\bar {xy}$  values we calculate from the training data 

```python
from mpl_toolkits.mplot3d import Axes3D
        
n = len(lindata.X)

def calculateH(b0, b1):
    return n * (y_sq_mean + (b1**2)*x_sq_mean + 2*b1*b0*x_mean - 2*b0*y_mean - 2*b1*xy_mean + b0**2)

xvals = np.linspace(-100, 100, 1000)
yvals = np.linspace(-100, 100, 1000)
theta0, theta1 = np.meshgrid(xvals, yvals)
hvals = calculateH(theta0, theta1)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(theta0, theta1, hvals, color="orange")
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('H')
plt.show()
```

#### Finding the minimum 

```python
b0 = (x_mean * xy_mean - x_square_mean * y_mean)/(x_mean**2 - x_square_mean)
b1 = (x_mean * y_mean - xy_mean)/(x_mean**2 - x_square_mean)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(theta0, theta1, hvals, color="orange")
ax.scatter(b0, b1, calculateH(b0, b1), color='red', marker='o', s=50, label="minimum point")
ax.view_init(0, 0)
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('H')
plt.legend()
plt.show()
```

## Assignment - 1 : Part 6

```python
x = np.linspace(-50, 50, 1000)
y = np.linspace(-50, 50, 1000)

theta0, theta1 = np.meshgrid(x, y)

gvals = (theta0**2) - (theta1**2)
```

#### Plotting the g($\Theta $0, $\Theta $1) function

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.plot_surface(theta0, theta1, gvals, color="orange")
ax.set_xlabel("theta0")
ax.set_ylabel("theta1")
ax.set_zlabel("G")
plt.show()
```

#### Showing the saddle point

```python
fig = plt.figure(figsize=plt.figaspect(0.5))

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(theta0, theta1, gvals, color="gold")
ax.scatter(0, 0, color="red", marker="o", s=100, label="saddle point")
ax.set_xlabel("theta0")
ax.set_ylabel("theta1")
ax.set_zlabel("G")

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(theta0, theta1, gvals, color="gold")
ax.scatter(0, 0, color="red", marker="o", s=100, label="saddle point")
ax.set_xlabel("theta0")
ax.set_ylabel("theta1")
ax.set_zlabel("G")
ax.view_init(0, 0)

plt.legend()
plt.show()
```
